```python
# Install Gradio if not already installed
!pip install gradio pandas

import gradio as gr
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/nutrition app/Diets1.csv'  # Replace with the correct path to your CSV file
data = pd.read_csv(file_path)

# Function to generate the disease dropdown menu
def get_disease_list():
    return data['Disease'].unique().tolist()

# Function to get the recommended diet for a selected disease
def get_diet(disease):
    diet = data.loc[data['Disease'] == disease, 'Diet'].values
    if diet:
        # Split diet items by commas (assuming they are comma-separated) and format them line-by-line
        diet_items = diet[0].split(',')  # Modify the delimiter if needed
        formatted_diet = "Recommended Diet:\n" + "\n".join(f"- {item.strip()}" for item in diet_items)
        return formatted_diet
    else:
        return "No recommended diet found for this disease."

# Gradio Interface
def disease_diet_app(disease):
    return get_diet(disease)

# Create the dropdown menu dynamically
disease_list = get_disease_list()

# Gradio app setup
interface = gr.Interface(
    fn=disease_diet_app,
    inputs=gr.Dropdown(choices=disease_list, label="Select a Disease (Medical Diet)"),
    outputs="text",
    title="Disease-Specific Diet Recommendations",
    description="Select a disease from the dropdown menu to view the recommended diet in line-by-line format.",
)

# Launch the app
interface.launch()
```
