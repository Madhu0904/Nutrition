```python
import gradio as gr
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/nutrition app/food1.csv.xlsx'  # Ensure the correct file path
food_data = pd.read_excel(file_path)

# Function to classify BMI
def classify_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Healthy"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

# Function to suggest food items based on diet type and goal
def suggest_food_items(diet_type, diet_goal):
    if diet_goal == "Weight Loss":
        # Low-calorie options
        filtered_items = food_data[(food_data['veg'] == diet_type) & (food_data['Calories'] < 200)]
    elif diet_goal == "Weight Gain":
        # High-calorie and protein-rich options
        filtered_items = food_data[(food_data['veg'] == diet_type) & (food_data['Calories'] > 200) & (food_data['Proteins'] > 5)]
    else:  # Healthy
        # Balanced options (moderate calories, proteins, and fats)
        filtered_items = food_data[(food_data['veg'] == diet_type) & (food_data['Calories'].between(100, 300)) & (food_data['Fats'] < 10)]

    return filtered_items['Food_items'].tolist()

# Main function for diet recommendation
def diet_recommendation(age, diet_type, weight, height, diet_goal):
    # Map diet type to binary values
    diet_type_binary = 1 if diet_type == "Vegetarian" else 0

    # BMI Calculation
    bmi = weight / ((height / 100) ** 2)
    bmi_status = classify_bmi(bmi)

    # Suggest Food Items
    food_suggestions = suggest_food_items(diet_type_binary, diet_goal)

    # Format Output
    result = (
        f"Age: {age}\n"
        f"Diet Type: {diet_type}\n"
        f"Weight: {weight} kg\n"
        f"Height: {height} cm\n\n"
        f"Your Body Mass Index (BMI) is: {bmi:.2f}\n"
        f"According to your BMI, you are: {bmi_status}\n"
        "####################\n"
        "SUGGESTED FOOD ITEMS:\n"
    )
    for item in food_suggestions:
        result += f"- {item}\n"

    result += "\nThank You for taking our recommendations. :)"
    return result

# Gradio App Setup
interface = gr.Interface(
    fn=diet_recommendation,
    inputs=[
        gr.Number(label="Age"),
        gr.Radio(["Vegetarian", "Non-Vegetarian"], label="Diet Type"),
        gr.Number(label="Weight (kg)"),
        gr.Number(label="Height (cm)"),
        gr.Radio(["Weight Loss", "Weight Gain", "Healthy"], label="Diet Plan Type"),
    ],
    outputs="text",
    title="Diet Recommendation System",
    description="Enter your details to get personalized diet recommendations based on the provided dataset.",
)

# Launch the app
interface.launch()
```
